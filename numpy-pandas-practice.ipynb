# Pandas Basics

## Creating DataFrame

```python
data = {
    "Name": ["Anil", "Rahul", "Priya", "Sneha"],
    "Marks": [85, 90, 88, 92],
    "City": ["Delhi", "Mumbai", "Pune", "Delhi"]
}

df = pd.DataFrame(data)

df
```

## Basic Functions

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Selecting Columns

```python
df["Name"]
```

```python
df[["Name", "Marks"]]
```

## Filtering Data 

```python
high_marks = df[df["Marks"] > 85]
high_marks
```

## Adding New Column

```python
df["Bonus"] = df["Marks"] + 5
df
```

## GroupBy Operations

```python
grouped = df.groupby("City")["Marks"].mean()
grouped
```

## Sorting Data

```python
sorted_df = df.sort_values(by="Marks", ascending=False)
sorted_df
```

## Handling Missing Values

```python
df.loc[2, "Marks"] = np.nan
df
```

### Check missing:

```python
df.isnull().sum()
```

### Fill missing:

```python
df["Marks"].fillna(df["Marks"].mean(), inplace=True)
df
```

## CSV File Operations

### Save CSV

```python
df.to_csv("student_data.csv", index=False)
```

### Load CSV

```python
new_df = pd.read_csv("student_data.csv")
new_df.head()
```
